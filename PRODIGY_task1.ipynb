```python
import pandas as pd  # for handling data
import matplotlib.pyplot as plt  # for making charts
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv', skiprows=4)
```

```python
data = df[['Country Name', '2020']]  # choose year
data = data.dropna()  # remove missing values
```

```python
plt.figure(figsize=(12,6))  # size of chart
plt.hist(data['2020'], bins=20, color='skyblue', edgecolor='black')  # create histogram

# Add labels and title
plt.title('Histogram: Population Distribution by Country (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
import pandas as pd

df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv', skiprows=4)
df['Indicator Name'].unique()[:20]  # See first 20 indicator types
```

```python
# Filter only female age 5-9 population
df_female_5_9 = df[df['Indicator Name'] == 'Population ages 5-9, female']

# Keep relevant columns
df_female_5_9 = df_female_5_9[['Country Name', '2020']].dropna()

# Preview
df_female_5_9.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.hist(df_female_5_9['2020'], bins=20, color='lightcoral', edgecolor='black')

plt.title('Population of Females Aged 59 by Country (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
