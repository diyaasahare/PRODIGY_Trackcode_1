```python
import pandas as pd  # for handling data
import matplotlib.pyplot as plt  # for making charts
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_2590.csv', skiprows=4)
```

```python
data = df[['Country Name', '2020']]  # choose year
data = data.dropna()  # remove missing values
```

```python
plt.figure(figsize=(12,6))  # size of chart
plt.hist(data['2020'], bins=20, color='skyblue', edgecolor='black')  # create histogram

# Add labels and title
plt.title('Histogram: Population Distribution by Country (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
